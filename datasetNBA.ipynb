H0: a media de falta dos jogadores da NBA é =0


H1: a media de falta dos jogares da nba é != 0

Interpretação: podemos afirmar que a media de faltas é significante pois o p-val 0<0.05, logo conseguimos rejeitar a hipotese nula chegando que a media dos jogadores é diferente de 0

```python
filtro1 = datasetNBA['GSCat'] == 'Titular'
filtro2 = datasetNBA['GSCat'] == 'Reserva'
Titular = datasetNBA[filtro1]
Reserva = datasetNBA[filtro2]
```

```python
[[Titular['PF'].mean(),Reserva['PF'].mean()]],[[Titular['PF'].std(),Reserva['PF'].std()]]
```

H0: A diferença na variaciancia das faltas cometidas por jogadores titulares e reservas não é significativa

---
H1: A diferença na variaciancia das faltas cometidas por jogadores titulares e reservas é significativa


```python
stats.bartlett(Titular['PF'],Reserva['PF'])
```

Temos um p-valor >0.05, logo podemos aceitar a hipotese nula que nos diz que a diferença entre a variancia das faltas cometidas dos jogadores titulares e reservas não é significativa

H0: a media de faltas dos jogadores titulares não é significativamente maior que a media de faltas dos jogadores reservas

---

H1: a media de faltas dos jogadores titulares é significativamente maior que a media de falta dos jogadores reservas

```python
ttest(Titular['PF'],Reserva['PF'],paired=False,alternative='greater')
```

Nós temos um valor menor que 0.05 para o p-val, logo rejeitamos a hipotese numa da qual assumimos como que a media de faltas dos jogadores titulares não é significativamente maior que a media de falta dos jogadores reservas

O eFG% (Percentual de Arremesso Eficaz) é uma formula no basquete que mede a eficiência de um jogador nos arremessos, considerando que arremessos de três pontos por serem mais dificeis possuem um peso maior. É parecido com o FG% (percentual de acerto), mas o eFG% dá um peso extra aos arremessos de três pontos, refletindo melhor o impacto real desses pontos no jogo.


sendo sua formula:

eFG% = (FG + (0.5 * 3P)) / FGA

```python
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=0.15, random_state=30)
```

```python
models = []
models.append(LogisticRegression(solver='liblinear', multi_class='ovr'))
models.append(LinearDiscriminantAnalysis())
models.append(KNeighborsClassifier())
models.append(DecisionTreeClassifier())
models.append(GaussianNB())
models.append(SVC(gamma='auto'))
```

```python
nome = ['LR','LDA','KNN','DT','NB','SVM']
```

```python
print(X.shape)  # Deve retornar (n_amostras, n_características)
print(Y.shape)
```

```python
results = [] #lista com os resultados de cada modelo
acuracia=[]
erro=[]
n=0

for model in models:
   Models_results = cross_val_score(model, X, Y, cv=5, scoring='accuracy',error_score='raise')

   results.append(Models_results)
   msg = "%s: %f (%f)" % (nome[n],round(Models_results.mean(),2), Models_results.std())
   acuracia.append(round(Models_results.mean(),2))
   erro.append(Models_results.std())
   n=n+1
   print(msg)

#scoring = 'accuracy' #Métrica Utiliza Para Avaliar os Modelos
```

```python
n = 1
resSilhouette = []
maxGrupo = []
components = X_scale.shape[1]

while n <= components:
    modelPca = PCA(n_components=n, random_state=43)
    modelPca.fit(X_scale)
    reduct_Diabets = modelPca.transform(X_scale)
    print('A variação do modelo quando utiliza {} componente é explicada em {}'.format(n, modelPca.explained_variance_ratio_.sum() * 100))

    clusters = [3, 4, 5, 6, 7, 8, 9, 10]  # Número de clusters a serem testados
    scoresSilhouette = []
    grupo = []
    maxsilhouete = 0.20
    maxcluster = 0

    for k in clusters:
        kmeans = KMeans(n_clusters=k, random_state=30).fit(reduct_Diabets)
        silhouette = silhouette_score(reduct_Diabets, kmeans.labels_)
        scoresSilhouette.append(silhouette)
        grupo.append(k)
        if ((silhouette > maxsilhouete) and ((modelPca.explained_variance_ratio_.sum() * 100) > 80)):
            maxsilhouete = silhouette
            maxcluster = k

    # Definir manualmente 10 clusters quando n == 10
    if n == 10:
        maxcluster = 10
        print(f'Número de clusters fixado em {maxcluster} para 10 componentes principais.')

    resSilhouette.append(maxsilhouete)
    print('O maior silhouette do grupo é {} com {} grupos'.format(max(scoresSilhouette), grupo[scoresSilhouette.index(max(scoresSilhouette))]))

    plt.figure(figsize=(12, 6))
    plt.plot(clusters, scoresSilhouette, marker='o')
    plt.title('Silhouette Score For {} PCA Component'.format(n))
    plt.xlabel('Número de Grupos')
    plt.ylabel('Silhouette Score')
    plt.show()

    scoresSilhouette = []
    maxsilhouete = 0.20
    maxcluster = 0
    n += 1

print(min(resSilhouette))
```
